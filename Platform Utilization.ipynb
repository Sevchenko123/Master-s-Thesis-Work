```python
import numpy as np
import pandas as pd
import seaborn as sns
sns.set_style('whitegrid')
import matplotlib as mpl
import matplotlib.pyplot as plt
from pandas import Series,DataFrame
from numpy.random import randn
from scipy import stats
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
```

```python
# Active harassers (1026) before and after block conversations
beforeblock = pd.read_csv('beforeblockactivehar.csv')
afterblock = pd.read_csv('afterblockactivehar.csv')
b = beforeblock['Freq']
a = afterblock['Freq']
```

```python
violin = sns.violinplot(b,inner='stick',vert=True)
violin.set(ylabel='Conversations')
sns.plt.title('Number of Conversations by Active Harassers before block')
average = b.mean()
std_dev = b.std()
print('The average conversations before block were %.2f with standard deviation %.2f for Active Harassers'%(average,std_dev))
```

```python
violin = sns.violinplot(a,inner='stick',vert=True)
violin.set(ylabel= 'Conversations')
sns.plt.title('Number of Conversations by Active Harassers after block')
average = a.mean()
std_dev = a.std()
print('The average conversations after block were %.2f with standard deviation %.2f for Active Harassers'%(average,std_dev))
```

```python
box = sns.boxplot(b,whis=np.inf,vert=True,palette='Set1')
box.set(ylabel='Conversations')
sns.plt.title('Number of Conversations by Active Harassers before block')
median = b.median()
print('The median number of conversations before block was %.2f for Active Harassers'%(median))
```

```python
box = sns.boxplot(a,whis=np.inf,vert=True,palette='Set1')
box.set(ylabel='Conversations')
sns.plt.title('Number of Conversations by Active Harassers after block')
median = a.median()
print('The median number of conversations after block was %.2f for Active Harassers'%(median))
```
